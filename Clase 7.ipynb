# Nueva forma de ordenar

Estrucura nueva: introducing *Heap*.

Tiene la ventaja de que se puede preguntar a la estructura cuál es el dato más prioritario de forma óptima. 

### Priority queue

- Insertar un dato con prioridad dada
- Extraer el dato con mayor prioridad
- (Idealmente): cambiar la prioridad de un dato ya insertado

Quizás no necesitamos mantener un orden total, pero sí mantener el elemento más prioritario a mano. Puede ser un orden parcial. 

¿Estructura recursiva? 
    - Si sigue el mismo patrón, la búsqueda también puede ser recursiva $\rightarrow$ más simple.
    - La implementación de la estructura puede ser más simple.

## Heap binario

A diferencia del árbol binario de búsqueda, que intenta mantener el orden (hijo izq $\leq$ padre $\leq$ hijo der), un `heap binario` (dependiendo si es max heap o min heap), mantiene al padre como mayor o menor elemento, no el del medio.
La diferencia es que tiene el elemento más prioritario en la raíz. Cada hijo de la raíz es, por sí mismo, un heap binario. Si tiene $h$ niveles (altura), puede almacenar hasta $2^h$ datos. Luego, la altura es $\O{\log n}$ si tengo $n$ datos.

¿Cómo podemos garantizar que llenamos el Heap de forma balanceada, con tal de minimizar la altura?

Si podemos suponer una cantidad máxima de datos que pueden estar en el Heap simultáneamente, puedo saber de antemano los niveles que requeriremos.
`OJO: por nivel, no se mantiene el orden de izquierda a derecha`.

```
                  31
             29         19
           7     23   13  17
          5 3   11
```
Se traduce a

$ \{ 31, 29, 19, 7, 23, 13, 17, 5, 3, 11 \}$

Si una hoja tiene un solo hijo, entonces será el último elemento del `array`.

¿Y si tengo que agregar un número mayor a 23?
Tenemos que agregar hijos desde izquierda a derecha.

Si hago extract(0), tomo el último y lo pongo primero. Esto, dado que no puedo dejar sin raíz. Pero esto también puede dejar desordenado, porque el 11 < 29 y también 11 < 19. Debo comparar con los hijos.

```
extract(H):
  i = última celda no vacía # (11 en el ejemplo)
  best = H[0]
  H[1] = H[i]
  H[i] = NULL
  siftdown(H, 1)
  return best
```
$ \{ 31, 29, 19, 7, 23, 13, 17, 5, 3, 11 \}$

$ \{ , 29, 19, 7, 23, 13, 17, 5, 3, 11 \}$

$ \{ 11, 29, 19, 7, 23, 13, 17, 5, 3, NULL \}$

Ahora, puedo intercambiar el 11 con el 29 (porque es el mayor), quedando este como raíz.

$ \{ 29, 11, 19, 7, 23, 13, 17, 5, 3, NULL \}$

¿Esto es un heap? No, porque el 11 > 23 (hijo). Se enroca:

$ \{ 29, 23, 19, 7, 11, 13, 17, 5, 3, NULL \}$

```
siftdown(H, i):
  if i tiene hijos:
    i' = el hijo de i con más prioridad
    if  H[i'] > H[i]: # hay un hijo con más prioridad que el padre
      H[i'] <-> H[i]
      siftdown(H, i')
```

Ahora, debemos insertar elementos:
```
insert(H, e):
  i = primera celda en blanco de H
  H[i] = e
  siftup(H, e)
```

# HeapSort
1. Convertir A en un `max heap` con los datos
2. Definir una secuencia ordenada, B, inicialmente vacía
3. Extraer el mayor dato $x$ de A e insertarlo en la última posición vacía de B
4. Si quedan elementos en A, volver a 2.

A diferencia de `MergeSort`, `HeapSort` es _inplace_, no requiere memoria adicional. B va a estar al principio y luego dejamos A, y cada vez que saquemos el primero de A, lo agregamos a B al final.

Cuando esté construido el Heap, voy sacando el primero (más prioridad) y lo voy agregando
